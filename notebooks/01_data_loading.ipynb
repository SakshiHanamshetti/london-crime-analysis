```python
# Loading Script
import pandas as pd
import glob

# This finds every CSV file inside all 12 month folders
file_paths = glob.glob("/content/drive/MyDrive/london-crime-analysis/data/raw/**/*.csv", recursive=True)

print(f"Found {len(file_paths)} files")
```

```python
# Read each CSV file and stack them all into one big table
df = pd.concat([pd.read_csv(f) for f in file_paths], ignore_index=True)

print(f"Total rows: {len(df)}")
print(f"Total columns: {len(df.columns)}")
print(df.columns.tolist())
```

```python
# How many nulls in each column?
print(df.isnull().sum())
```

```python
# Drop the Context column - completely empty, useless
df.drop(columns=['Context'], inplace=True)

# Drop rows where location is unknown (LSOA name is null)
df.dropna(subset=['LSOA name'], inplace=True)

# Fill missing outcome with 'Not recorded'
df['Last outcome category'].fillna('Not recorded', inplace=True)

# Convert Month to proper datetime format
df['Month'] = pd.to_datetime(df['Month'])

# Confirm the result
print(f"Rows after cleaning: {len(df)}")
print(f"\nNull check after cleaning:")
print(df.isnull().sum())
```

```python
df.drop(columns=['Crime ID'], inplace=True)
```

```python
# Save cleaned data as a single CSV
df.to_csv("/content/drive/MyDrive/london-crime-analysis/data/cleaned_crime_data.csv", index=False)

print("Saved successfully!")
print(f"Final dataset: {len(df):,} rows, {len(df.columns)} columns")
```
